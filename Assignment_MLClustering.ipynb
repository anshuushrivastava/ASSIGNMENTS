##Theoritical Questions

Que1: What is unsupervised learning in the context of machine learning?

Ans1: Unsupervised learning is a machine learning approach where the model is trained on data without labeled outputs. It aims to discover hidden patterns or structures in the input data. Common tasks include clustering and dimensionality reduction, enabling insights without predefined categories or outcomes.

Que2: How does K-Means clustering algorithm work?

Ans2: K-Means clustering partitions data into k distinct clusters. It begins with randomly initializing k centroids, assigns each data point to the nearest centroid, then recalculates centroids based on the assigned points. This process repeats iteratively until the centroids stabilize or a set number of iterations is reached.

Que3: Explain the concept of a dendrogram in hierarchical clustering.

Ans3: A dendrogram is a tree-like diagram that visualizes the arrangement of clusters formed by hierarchical clustering. It shows how individual data points merge into clusters step by step. The height of the branches indicates the distance or dissimilarity between the merged clusters.

Que4: What is the main difference between K-Means and Hierarchical Clustering?

Ans4: The main difference is in the approach: K-Means requires a pre-specified number of clusters and uses an iterative optimization method, whereas Hierarchical Clustering builds a tree of clusters without requiring the number of clusters in advance.

Que5: What are the advantages of DBSCAN over K-Means?

Ans5: DBSCAN does not require the number of clusters to be specified beforehand, can find clusters of arbitrary shapes, and is robust to noise and outliers. In contrast, K-Means assumes spherical clusters and can be sensitive to outliers.

Que6: When would you use Silhouette Score in clustering?

Ans6: Silhouette Score is used to evaluate the quality of clustering by measuring how well each point fits within its cluster compared to others. It is particularly helpful for determining the optimal number of clusters and for comparing clustering algorithms.

Que7: What are the limitations of Hierarchical Clustering?

Ans7: Hierarchical Clustering can be computationally expensive for large datasets, lacks flexibility once the dendrogram is built, and can be sensitive to noise and outliers. Additionally, it does not scale well with dimensionality or data size.

Que8: Why is feature scaling important in clustering algorithms like K-Means?

Ans8: Feature scaling ensures that all features contribute equally to distance calculations. Without scaling, variables with larger ranges can disproportionately influence cluster assignments, leading to biased and inaccurate clustering results.

Que9: How does DBSCAN identify noise points?

Ans9: DBSCAN identifies noise points as those that do not belong to any cluster. Specifically, a point is considered noise if it has fewer neighbors than the minimum required (MinPts) within a specified radius (ε), making it neither a core nor a border point.

Que10: Define inertia in the context of K-Means.

Ans10: Inertia is the sum of squared distances between each data point and its assigned cluster centroid. It measures the compactness of the clusters. Lower inertia indicates tighter clusters, but it should be balanced with model complexity to avoid overfitting.

Que11: What is the elbow method in K-Means clustering?

Ans11: The elbow method helps determine the optimal number of clusters by plotting inertia against the number of clusters. The "elbow" point—where the rate of decrease sharply changes—suggests a balance between compactness and simplicity.

Que12: Describe the concept of "density" in DBSCAN.

Ans12: In DBSCAN, density refers to the number of points within a given radius (ε) around a point. A region is considered dense if it contains at least MinPts points. This helps identify clusters based on the density of data points rather than distance alone.

Que13: Can hierarchical clustering be used on categorical data?

Ans13: Yes, hierarchical clustering can be adapted for categorical data by using appropriate distance measures like Hamming or Jaccard distance. However, care must be taken to choose linkage methods and metrics suited for non-numeric data.

Que14: What does a negative Silhouette Score indicate?

Ans14: A negative Silhouette Score indicates that a data point may be assigned to the wrong cluster. It implies that the point is closer to points in another cluster than to those in its own, suggesting poor clustering quality.

Que15: Explain the term "linkage criteria" in hierarchical clustering.

Ans15: Linkage criteria determine how the distance between clusters is calculated when merging them. Common methods include single linkage (minimum distance), complete linkage (maximum distance), and average linkage (mean distance between all pairs).

Que16: Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Ans16: K-Means assumes clusters are spherical and equally sized, which makes it struggle with datasets containing clusters of varying sizes, densities, or non-convex shapes. It may incorrectly merge or split clusters, leading to inaccurate results.

Que17: What are the core parameters in DBSCAN, and how do they influence clustering?

Ans17: The two main parameters are ε (epsilon) and MinPts. Epsilon defines the neighborhood radius, and MinPts specifies the minimum number of points required to form a dense region. Together, they determine cluster shape, density sensitivity, and noise detection.

Que18: How does K-Means++ improve upon standard K-Means initialization?

Ans18: K-Means++ improves K-Means by selecting initial centroids in a smarter way—spreading them out to reduce the likelihood of poor clustering results. This leads to faster convergence and often better final clusters compared to random initialization.

Que19: What is agglomerative clustering?

Ans19: Agglomerative clustering is a bottom-up approach in hierarchical clustering. It starts by treating each data point as a single cluster and successively merges the closest pairs of clusters based on a distance metric until all points are grouped into one cluster.

Que20: What makes Silhouette Score a better metric than just inertia for model evaluation?

Ans20: Silhouette Score evaluates how well-separated and cohesive the clusters are, combining both inter-cluster and intra-cluster distances. Inertia only measures compactness, which can be misleading when comparing clusters of different shapes or densities.


##Practical Answers

Que21:  Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

```python
#Ans21:
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42).fit(X)

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("K-Means Clustering on make_blobs")
plt.show()
```

Que22: Load the Iris dataset and use Agglomerative Clustering to group the data into 3 clusters. Display the first 10 predicted labels.

Que29: Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
#Ans29:
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=500, centers=4, random_state=42)
scores = []
for k in range(2, 6):
    km = KMeans(n_clusters=k, random_state=42).fit(X)
    scores.append(silhouette_score(X, km.labels_))

plt.bar(range(2, 6), scores)
plt.title("Silhouette Scores for k=2 to 5")
plt.xlabel("k")
plt.ylabel("Score")
plt.show()
```

Que31: Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
#Ans31:
from matplotlib.colors import ListedColormap
import numpy as np

X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap='Pastel1')
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, edgecolors='k')
plt.title("K-Means with Decision Boundaries")
plt.show()
```
