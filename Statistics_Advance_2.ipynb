Que1: What is hypothesis testing in statistics?

Ans1: Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves two hypotheses:

- Null hypothesis (\( H_0 \)): Assumes no effect or difference.
- Alternative hypothesis (\( H_1 \)): Assumes there is an effect or difference.

A test statistic is calculated from the sample, and a p-value is used to assess the evidence against \( H_0 \). If the p-value is less than a chosen significance level (e.g., 0.05), \( H_0 \) is rejected. This process helps determine whether the observed results are statistically significant.


Que2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans2: The null hypothesis (\( H_0 \)) assumes there is no effect or difference, while the alternative hypothesis (\( H_1 \)) suggests there is an effect or difference.

- \( H_0 \): Default assumption (e.g., no change)
- \( H_1 \): Opposes \( H_0 \) (e.g., some change)
- If evidence is strong (p-value ≤ α), we reject \( H_0 \) in favor of \( H_1 \).  
- If not, we do not reject \( H_0 \).

Que3: What is the significance level in hypothesis testing, and why is it important?

Ans3: The significance level (denoted as α) is the threshold probability used in hypothesis testing to decide whether to reject the null hypothesis.

- It represents the risk of making a Type I error, i.e., rejecting a true null hypothesis.
- Common values: 0.05 (5%), 0.01 (1%), etc.
- If p-value ≤ α, we reject \( H_0 \) (result is statistically significant).
- If p-value > α, we do not reject \( H_0 \).

Que4: What does a P-value represent in hypothesis testing?

Ans4: A p-value in hypothesis testing represents the probability of obtaining results as extreme as or more extreme than the observed data, assuming the null hypothesis (\( H_0 \)) is true.

- Low p-value (≤ α) → strong evidence against \( H_0 \) → reject \( H_0 \).
- High p-value (> α) → weak evidence against \( H_0 \) → do not reject \( H_0 \).

Que5: How do you interpret the P-value in hypothesis testing?

Ans5: Interpreting the P-value in hypothesis testing:

- P-value ≤ α (e.g., 0.05):  
  There is strong evidence against the null hypothesis (\( H_0 \)), so you reject \( H_0 \).  
  → The result is statistically significant.

- P-value > α:  
  There is not enough evidence to reject \( H_0 \).  
  → The result is not statistically significant.

Que6: What are Type 1 and Type 2 errors in hypothesis testing

Ans6: When making a decision in hypothesis testing, two types of errors can occur:

Type I Error (False Positive):
Definition: Rejecting the null hypothesis H0 when it is actually true.
Symbol: α (significance level)
Example: Concluding a drug works when it actually doesn't.

Type II Error (False Negative):
Definition: Failing to reject the null hypothesis H0 when it is actually false.
Symbol: β
Example: Concluding a drug doesn't work when it actually does.

Que7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing

Ans7: In hypothesis testing, the choice between a one-tailed and two-tailed test depends on the direction of the research question.

One-Tailed Test:
- Tests for an effect in one specific direction (either greater than or less than).
- Alternative Hypothesis: H1 : µ > µ0  or H1 : µ < µ0  
- Usually its used only when one is interested in detecting an increase or a decrease.

Two-Tailed Test:
- Tests for an effect in both directions (either greater or less).
- Alternative Hypothesis: H1:μ ≠ μ0
- Usually its used to check any difference (positive or negative).

Que9: How do you calculate the Z-score, and what does it represent in hypothesis testing?

Ans9: The Z-score measures how many standard deviations a sample statistic (like the sample mean) is from the population mean, under the assumption that the null hypothesis is true.
Formula (for sample mean): Z = ( x̄ - μ ) / (σ / √n)

Where:
- x̄ = sample mean  
- μ= population mean (under \( H_0 \))  
- σ = population standard deviation  
- n= sample size  

- The Z-score shows how far the sample result is from the expected population mean in standard deviation units.
- It is used to find the p-value by comparing it to the standard normal distribution.
- A large absolute Z-score (e.g., > 1.96 for α = 0.05 in a two-tailed test) indicates the result is unlikely under \( H_0 \) → evidence to reject \( H_0 \).

Que18: What is the Chi-square goodness of fit test, and how is it applied?

Ans18: The Chi-square goodness of fit test is a statistical test used to determine whether a set of observed categorical data matches an expected distribution.

To check if the observed frequencies in categories are consistent with expected frequencies based on a certain hypothesis.

Working-
1. State Hypotheses:
- Null hypothesis H0: The observed data fits the expected distribution.

- Alternative hypothesis H1: The observed data does not fit the expected distribution.

2. Calculate Test Statistic: χ² = Σ [(Oᵢ - Eᵢ)² / Eᵢ]
Where:
Oᵢ = Observed frequency in category i
Eᵢ = Expected frequency in category i

3. Determine Degrees of Freedom (df):
df = Number of categories - 1

4. Compare χ²  with the critical value from the chi-square distribution table OR use the p-value.

5. Decision:
- If χ² is greater than the critical value → Reject H0
- Otherwise → Fail to reject H0

```python
#Ans3:
def bernoulli_pmf(x, p):
    if x not in [0, 1]:
        return 0.0
    return p if x == 1 else (1 - p)

# Example usage:
print("P(X=0):", bernoulli_pmf(0, 0.7))
print("P(X=1):", bernoulli_pmf(1, 0.7))
```

Que9:  Write a Python function to calculate Z-scores from a dataset and plot them


```python
#Ans9:
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Calculating Z-scores:-
    z_scores = (data - mean) / std

    # Plotting Z-scores:-
    plt.figure(figsize=(8, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title("Z-scores of the Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return z_scores

data = [10, 12, 15, 18, 20, 22, 25, 30, 28, 27]
calculate_and_plot_zscores(data)
```

Que18: Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution


```python
#Ans18:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    return (x - mean) / std_dev

def compare_to_standard_normal(z):
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, 0, 1)

    plt.plot(x_vals, y_vals, label="Standard Normal PDF", color='blue')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= z), color='red', alpha=0.3)

    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    prob = norm.cdf(z)
    print(f"Z-score: {z:.2f}")
    print(f"Probability that a value is ≤ {z:.2f}: {prob:.4f}")

x = 75           # Data point
mean = 70        # Population mean
std_dev = 10     # Population standard deviation

z_score = calculate_z_score(x, mean, std_dev)
compare_to_standard_normal(z_score)
```

Que19: Implement hypothesis testing using Z-statistics for a sample dataset.


```python
#Ans19:
import numpy as np
from scipy.stats import norm

sample_data = np.array([52, 55, 48, 50, 49, 53, 54, 51, 52, 50])

population_mean = 50
population_std = 3  # Assumed known for Z-test

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

alpha = 0.05

z_critical = norm.ppf(1 - alpha / 2)

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.2f}")
print(f"Critical Z-value = ±{z_critical:.2f}")

if abs(z_stat) > z_critical:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

Que20: Create a confidence interval for a dataset using Python and interpret the result

```python
#Ans20:
import numpy as np
from scipy.stats import norm

data = np.array([52, 55, 48, 50, 49, 53, 54, 51, 52, 50])

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

confidence = 0.95
alpha = 1 - confidence

z_critical = norm.ppf(1 - alpha / 2)

margin_of_error = z_critical * (sample_std / np.sqrt(n))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"{int(confidence * 100)}% Confidence Interval = ({lower_bound:.2f}, {upper_bound:.2f})")
```

Que5: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing


```python
#Ans5:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for a one-tailed test (right-tailed)
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # Type II error (P(fail to reject H0 | H1 true))
    beta = norm.cdf((x_crit - mu1) / se)

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y_null = norm.pdf(x, mu0, se)
    y_alt = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))

    plt.plot(x, y_null, label="Null Hypothesis (H₀)", color="blue")
    plt.fill_between(x, y_null, where=(x > x_crit), color='red', alpha=0.3, label="Type I Error (α)")

    plt.plot(x, y_alt, label="Alternative Hypothesis (H₁)", color="green")
    plt.fill_between(x, y_alt, where=(x <= x_crit), color='orange', alpha=0.3, label="Type II Error (β)")

    plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    print(f"Critical Value (Sample Mean): {x_crit:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30)
```

Que9: Write a Python function to calculate the confidence interval for a sample mean and explain its significance


```python
#Ans9:
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    margin = t_crit * std_err

    lower = mean - margin
    upper = mean + margin

    return mean, (lower, upper)

np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=10, size=30)
mean, (low, high) = confidence_interval(sample_data, confidence=0.95)

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({low:.2f}, {high:.2f})")
```

Que10: Write a Python program to calculate the margin of error for a given confidence level using sample data


```python
#Ans10:
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)

    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    moe = t_crit * standard_error
    return moe

np.random.seed(2)
sample_data = np.random.normal(loc=50, scale=8, size=25)
moe = margin_of_error(sample_data, confidence=0.95)

print(f"Sample Size = {len(sample_data)}")
print(f"Sample Mean = {np.mean(sample_data):.2f}")
print(f"95% Margin of Error = ±{moe:.2f}")
```

Que11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.


```python
#Ans11:
import numpy as np

def bayesian_update(priors, likelihoods):
    priors = np.array(priors)
    likelihoods = np.array(likelihoods)

    unnormalized_posterior = likelihoods * priors

    evidence = np.sum(unnormalized_posterior)
    posterior = unnormalized_posterior / evidence
    return posterior

# Example: Disease vs No Disease
hypotheses = ['Disease', 'No Disease']
priors = [0.01, 0.99]                   # Prior probabilities
likelihoods = [0.99, 0.05]              # Likelihood of positive test: P(E|Disease), P(E|No Disease)

posterior = bayesian_update(priors, likelihoods)

for h, p in zip(hypotheses, posterior):
    print(f"P({h} | Positive Test) = {p:.4f}")
```

Que16: Implement an F-test using Python to compare the variances of two random samples


```python
#Ans16:
import numpy as np
from scipy.stats import f

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1 with variance ~100
sample2 = np.random.normal(loc=55, scale=15, size=100)  # Sample 2 with variance ~225

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

if var1 > var2:
    f_stat = var1 / var2
    dfn = len(sample1) - 1  # degrees of freedom numerator
    dfd = len(sample2) - 1  # degrees of freedom denominator
else:
    f_stat = var2 / var1
    dfn = len(sample2) - 1
    dfd = len(sample1) - 1

p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("\n✅ Reject the null hypothesis: Variances are significantly different.")
else:
    print("\n❌ Fail to reject the null hypothesis: No significant difference in variances.")
```

Que19: Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA


```python
#Ans19:
import numpy as np
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups):
    all_data = [np.array(g) for g in groups]
    labels = [f'Group {i+1}' for i in range(len(groups))]

    print("🔍 Checking Normality (Shapiro-Wilk Test):")
    for i, group in enumerate(all_data):
        stat, p = shapiro(group)
        print(f"  Group {i+1}: W = {stat:.4f}, p = {p:.4f}")
        if p > 0.05:
            print("   ✅ Data looks normally distributed.")
        else:
            print("   ❌ Data may not be normally distributed.")

    print("\n🔍 Checking Homogeneity of Variances (Levene's Test):")
    stat, p = levene(*all_data)
    print(f"  Levene Statistic = {stat:.4f}, p = {p:.4f}")
    if p > 0.05:
        print("   ✅ Variances are equal across groups.")
    else:
        print("   ❌ Variances are significantly different across groups.")

    print("\nℹ️ Independence is assumed to be met by study design.")

    plt.figure(figsize=(8, 5))
    sns.boxplot(data=all_data)
    plt.xticks(ticks=np.arange(len(labels)), labels=labels)
    plt.title("Boxplot of Groups")
    plt.ylabel("Values")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
#Ans20:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

np.random.seed(42)
n = 20

treatment = np.repeat(['A', 'B'], n * 2)
gender = np.tile(np.repeat(['Male', 'Female'], n), 2)

score = (
    np.random.normal(50, 5, n).tolist() +
    np.random.normal(55, 5, n).tolist() +
    np.random.normal(52, 5, n).tolist() +
    np.random.normal(60, 5, n).tolist()
)

data = pd.DataFrame({
    'Treatment': treatment,
    'Gender': gender,
    'Score': score
})

model = ols('Score ~ C(Treatment) + C(Gender) + C(Treatment):C(Gender)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

print("\nInterpretation:")
for factor in anova_table.index:
    p = anova_table.loc[factor, "PR(>F)"]
    result = "✅ Significant" if p < 0.05 else "❌ Not Significant"
    print(f"{factor}: p-value = {p:.4f} → {result}")

plt.figure(figsize=(8, 6))
sns.pointplot(x="Treatment", y="Score", hue="Gender", data=data, capsize=0.1, dodge=True, markers=["o", "s"])
plt.title("Interaction Plot: Treatment × Gender")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Que21: Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.


```python
#Ans21:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

dfn = 5  # degrees of freedom (numerator)
dfd = 20  # degrees of freedom (denominator)

x = np.linspace(0, 5, 500)
pdf = f.pdf(x, dfn, dfd)

plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label=f'F-distribution (dfn={dfn}, dfd={dfd})', color='purple')
plt.fill_between(x, 0, pdf, where=(x > f.ppf(0.95, dfn, dfd)), color='orange', alpha=0.5, label='Rejection Region (α=0.05)')
plt.axvline(f.ppf(0.95, dfn, dfd), color='red', linestyle='--', label='Critical value (α=0.05)')
plt.title("F-distribution and Critical Region for Hypothesis Testing")
plt.xlabel("F-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Que22: Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.


```python
#Ans22:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

np.random.seed(42)

group_A = np.random.normal(loc=50, scale=5, size=30)
group_B = np.random.normal(loc=55, scale=5, size=30)
group_C = np.random.normal(loc=60, scale=5, size=30)

data = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A']*30 + ['B']*30 + ['C']*30
})

f_stat, p_value = f_oneway(group_A, group_B, group_C)
print("F-statistic:", round(f_stat, 4))
print("P-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("\n✅ Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("\n❌ Fail to reject the null hypothesis: No significant difference between group means.")

plt.figure(figsize=(8, 5))
sns.boxplot(x='Group', y='Score', data=data, palette="Set2")
plt.title("Boxplot of Scores by Group (One-Way ANOVA)")
plt.ylabel("Score")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Que23: Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.


```python
#Ans23:
import numpy as np
from scipy import stats

np.random.seed(0)
sample_data = np.random.normal(loc=102, scale=10, size=50)  # mean=102, std=10, n=50

population_mean = 100

t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)

print("Sample Mean:", round(np.mean(sample_data), 2))
print("Population Mean (H₀):", population_mean)
print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("\n✅ Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("\n❌ Fail to reject the null hypothesis: No significant difference between sample and population means.")
```

Que24: Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.


```python
#Ans24:
import numpy as np
from scipy.stats import chi2

np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, SD=5, n=30

sigma0_squared = 25  # (SD = 5)

n = len(sample)
sample_var = np.var(sample, ddof=1)

chi_square_stat = (n - 1) * sample_var / sigma0_squared

alpha = 0.05
lower_crit = chi2.ppf(alpha / 2, df=n - 1)
upper_crit = chi2.ppf(1 - alpha / 2, df=n - 1)

p_value = 2 * min(
    chi2.cdf(chi_square_stat, df=n - 1),
    1 - chi2.cdf(chi_square_stat, df=n - 1)
)

print("Sample Variance:", round(sample_var, 4))
print("Chi-square Statistic:", round(chi_square_stat, 4))
print("Degrees of Freedom:", n - 1)
print("P-value:", round(p_value, 4))
print(f"Critical Region: χ² < {lower_crit:.2f} or χ² > {upper_crit:.2f}")

if chi_square_stat < lower_crit or chi_square_stat > upper_crit:
    print("\n✅ Reject H₀: The population variance is significantly different.")
else:
    print("\n❌ Fail to reject H₀: No significant difference in population variance.")
```

Que25: Write a Python script to perform a Z-test for comparing proportions between two datasets or groups


```python
#Ans25:
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

successes = np.array([45, 30])      # e.g., 45 successes out of 100, 30 out of 80
sample_sizes = np.array([100, 80])

z_stat, p_value = proportions_ztest(successes, sample_sizes)

print("Z-statistic:", round(z_stat, 4))
print("P-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("\n✅ Reject the null hypothesis: Proportions are significantly different.")
else:
    print("\n❌ Fail to reject the null hypothesis: No significant difference in proportions.")
```

Que26: Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results


```python
#Ans26:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=30)  # std = 10
data2 = np.random.normal(loc=55, scale=5, size=30)   # std = 5

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

f_stat = var1 / var2
df1 = len(data1) - 1
df2 = len(data2) - 1

if f_stat > 1:
    p_value = 2 * (1 - f.cdf(f_stat, df1, df2))
else:
    p_value = 2 * f.cdf(f_stat, df1, df2)

print("Variance of Data1:", round(var1, 4))
print("Variance of Data2:", round(var2, 4))
print("F-statistic:", round(f_stat, 4))
print("Degrees of freedom:", df1, "and", df2)
print("P-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("\n✅ Reject the null hypothesis: Variances are significantly different.")
else:
    print("\n❌ Fail to reject the null hypothesis: No significant difference in variances.")

plt.figure(figsize=(10, 5))
plt.hist(data1, bins=15, alpha=0.6, label='Dataset 1 (std=10)', color='skyblue')
plt.hist(data2, bins=15, alpha=0.6, label='Dataset 2 (std=5)', color='salmon')
plt.axvline(np.mean(data1), color='blue', linestyle='--', linewidth=1.5)
plt.axvline(np.mean(data2), color='red', linestyle='--', linewidth=1.5)
plt.title('Comparison of Distributions (F-Test for Variance)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Que27: Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
#Ans27:
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

np.random.seed(42)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=120, p=[0.1, 0.2, 0.15, 0.25, 0.2, 0.1])
unique, counts = np.unique(observed, return_counts=True)

expected = np.array([len(observed)/6] * 6)  # Equal expected count in each category

chi_stat, p_value = chisquare(f_obs=counts, f_exp=expected)

print("Observed counts:", counts)
print("Expected counts:", expected.astype(int))
print("Chi-square Statistic:", round(chi_stat, 4))
print("P-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("\n✅ Reject H₀: The data does not follow the expected distribution.")
else:
    print("\n❌ Fail to reject H₀: The data fits the expected distribution.")

labels = [str(i) for i in unique]
x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, counts, width, label='Observed', color='orange')
plt.bar(x + width/2, expected, width, label='Expected', color='steelblue')
plt.xticks(x, labels)
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
