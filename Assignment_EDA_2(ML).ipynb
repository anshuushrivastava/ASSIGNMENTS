Que6: How does the average selling price of cars vary by customer gender and annual income?

Que7: What is the distribution of car prices by region, and how does the number of cars sold vary by region?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
try:
    df = pd.read_csv("CarSale.csv")
    print(df.head())

    # Check if 'region' and 'selling_price' columns exist
    if 'region' in df.columns and 'selling_price' in df.columns:
        # Create side-by-side plots
        fig, ax = plt.subplots(1, 2, figsize=(14, 5))

        # Boxplot: Selling price by region
        sns.boxplot(x='region', y='selling_price', data=df, ax=ax[0])
        ax[0].set_title('Selling Price by Region')
        ax[0].tick_params(axis='x', rotation=45)

        # Bar plot: Number of cars sold by region
        df['region'].value_counts().plot(kind='bar', ax=ax[1], color='skyblue')
        ax[1].set_title('Cars Sold by Region')
        ax[1].set_ylabel('Number of Cars')
        ax[1].set_xlabel('Region')
        ax[1].tick_params(axis='x', rotation=45)

        # Layout and display
        plt.tight_layout()
        plt.show()
    else:
        print("Error: 'region' and/or 'selling_price' column not found in the dataset.")

except FileNotFoundError:
    print("Error: 'CarSale.csv' not found. Please check the file name and path.")
```

Que9: How do car prices vary based on the customer's annual income bracket?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
try:
    df = pd.read_csv("CarSale.csv")
    print(df.head())  # Display the first few rows

    # Check if required columns exist
    if 'annual_income' in df.columns and 'selling_price' in df.columns:

        # Remove rows with NaN values in relevant columns
        df = df.dropna(subset=['annual_income', 'selling_price'])

        # Create income brackets using qcut
        df['income_bracket'] = pd.qcut(df['annual_income'], q=4, labels=['Low', 'Mid-Low', 'Mid-High', 'High'])

        # Create the boxplot
        sns.boxplot(x='income_bracket', y='selling_price', data=df)
        plt.title('Car Price by Income Bracket')
        plt.xlabel('Income Bracket')
        plt.ylabel('Selling Price')
        plt.tight_layout()  # To ensure the plot is well-spaced
        plt.show()

    else:
        print("Error: 'annual_income' and/or 'selling_price' columns not found in the dataset.")

except FileNotFoundError:
    print("Error: 'CarSale.csv' not found. Please check the file name and path.")
```

Que12: Is there any seasonal trend in car sales based on the date of sale?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
try:
    df = pd.read_csv("CarSale.csv")
    print(df.head())  # Display the first few rows

    # Check if 'date_of_sale' column exists
    if 'date_of_sale' in df.columns:

        # Convert the 'date_of_sale' to datetime format
        df['date_of_sale'] = pd.to_datetime(df['date_of_sale'], errors='coerce')  # 'coerce' will turn invalid parsing into NaT

        # Handle rows with invalid 'date_of_sale' values
        df = df.dropna(subset=['date_of_sale'])

        # Extract the month from the 'date_of_sale' column
        df['month'] = df['date_of_sale'].dt.month

        # Group by month and count the number of sales
        sales_by_month = df.groupby('month').size()

        # Plot the sales by month
        sales_by_month.plot(kind='bar', title='Monthly Car Sales', color='skyblue')
        plt.xlabel('Month')
        plt.ylabel('Cars Sold')
        plt.xticks(rotation=0)  # Ensures months are readable
        plt.tight_layout()  # Adjust layout for better spacing
        plt.show()
    else:
        print("Error: 'date_of_sale' column not found in the dataset.")

except FileNotFoundError:
    print("Error: 'CarSale.csv' not found. Please check the file name and path.")
```

Que14: What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
try:
    df = pd.read_csv("CarSale.csv")
    print(df.head())  # Display the first few rows

    # Check if the required columns exist
    if all(col in df.columns for col in ['selling_price', 'engine_size', 'annual_income']):

        # Remove rows with NaN values in the relevant columns
        df = df.dropna(subset=['selling_price', 'engine_size', 'annual_income'])

        # Calculate the correlation matrix
        corr = df[['selling_price', 'engine_size', 'annual_income']].corr()

        # Create the heatmap
        sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
        plt.title('Correlation Matrix')
        plt.tight_layout()  # Adjust layout for better spacing
        plt.show()
    else:
        print("Error: One or more required columns ('selling_price', 'engine_size', 'annual_income') not found in the dataset.")

except FileNotFoundError:
    print("Error: 'CarSale.csv' not found. Please check the file name and path.")
```

Que15: How does the average car price vary across different car models and engine types?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
try:
    df = pd.read_csv("CarSale.csv")
    print(df.head())  # Display the first few rows

    # Check if the required columns exist
    if all(col in df.columns for col in ['model', 'engine_size', 'selling_price']):

        # Remove rows with NaN values in the relevant columns
        df = df.dropna(subset=['model', 'engine_size', 'selling_price'])

        # Group by 'model' and 'engine_size' and calculate the mean selling price
        avg_price_model_engine = df.groupby(['model', 'engine_size'])['selling_price'].mean().unstack()

        # Plot the stacked bar chart
        avg_price_model_engine.plot(kind='bar', stacked=True, figsize=(12,6), cmap='viridis')
        plt.title('Avg Car Price by Model and Engine Size')
        plt.ylabel('Avg Price')
        plt.xlabel('Car Model')
        plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
        plt.tight_layout()  # Adjust layout for better spacing
        plt.show()
    else:
        print("Error: One or more required columns ('model', 'engine_size', 'selling_price') not found in the dataset.")

except FileNotFoundError:
    print("Error: 'CarSale.csv' not found. Please check the file name and path.")
```
