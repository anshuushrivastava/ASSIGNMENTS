Que1:  What is the difference between a function and a method in Python?

Ans1:
#Function:
- A function is a block of code that is defined independently and can be called directly.
- It may or may not take parameters, and it returns a value (optional).
- Functions are defined using the def keyword.

```python
#Example of Function
def greet(name):
    return f"Welcome to the Python learning club, {name}!"
```

#Method
- A method is a function that is associated with an object (usually defined within a class) and is called on that object.
- Methods operate on the data (attributes) of the object and can modify the object’s state.
- Methods are called using the syntax object.method().

```python
#Example of Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

person = Person("Alice")
print(person.greet())        # Method call on an object
```

Que5: What are iterators in Python and how do they differ from iterables?

Ans5: Iterators:
- An iterator is an object that represents a stream of data and allows iteration over it, one element at a time.
- An iterator implements two methods:
  1. `__iter__()`: Returns the iterator object itself (required for an object to be an iterable).
  2. `__next__()`: Returns the next value from the iterator. When there are no more items, it raises the `StopIteration` exception.

Iterables:
- An iterable is any object that can return an iterator using the `__iter__()` method.
- Examples of iterables include lists, tuples, strings, and dictionaries.
- Iterables can be traversed by creating an iterator, which fetches elements one by one.

Key Difference:
- **Iterable**: An object that can return an iterator (like a list or tuple).
- **Iterator**: An object that performs the actual iteration using `__next__()`.



```python
#Example:
# Iterable
my_list = [1, 2, 3]

# Convert iterable to iterator
my_iterator = iter(my_list)

# Using the iterator to get elements
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
```

Que6: Explain the concept of generators in Python and how they are defined.

Ans6: Generators:
- A generator is a special type of iterator that is defined using a function with the yield keyword.
- Unlike normal functions that return a value once and exit, generators produce a sequence of values lazily, meaning they generate each value on the fly as needed, rather than storing the entire sequence in memory.
- Generators are memory-efficient and allow you to iterate over large datasets without loading everything into memory at once.

A generator is defined using a function that contains one or more yield statements.The yield keyword allows the function to return a value and pause its state. When next() is called on the generator, execution resumes from where it left off.

```python
#Example
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yield the current count and pause
        count += 1

# Creating a generator object
counter = count_up_to(5)

# Iterating over the generator
for num in counter:
    print(num)
```

Que7: What are the advantages of using generators over regular functions?

Ans7: Advantages of Using Generators Over Regular Functions in Python:

1. Memory Efficiency:
   Generators produce values one at a time and don’t store the entire sequence in memory, making them memory-efficient, especially for large datasets. Regular functions that return lists store all values at once.

2. Lazy Evaluation:
   Generators generate values only when needed, reducing unnecessary computations. Regular functions return all values at once, even if only part of them is required.

3. Performance:
   Generators improve performance by generating values on-demand and avoiding the overhead of creating large data structures. They are more efficient for iterative processing.

4. Simplified Code:
   Generators simplify iteration logic, as the `yield` statement automatically handles pausing and resuming function execution.

5. Infinite Sequences:
   Generators can handle infinite sequences (like Fibonacci numbers) because they generate values as needed, while regular functions that return lists cannot handle infinite data.

Generators are perfect for large or infinite datasets, improving memory usage and performance.

Que9: Explain the purpose and usage of the `map()` function in Python.

Ans9: The `map()` function in Python is used to apply a specific function to each item in an iterable (like a list, tuple, or string) and returns an iterator that yields the results. It is useful for transforming data in a concise way, especially when you want to perform the same operation on all elements of the iterable.
Purpose:
- To transform data by applying a function to each element in an iterable.
- Helps in performing repetitive tasks without explicitly writing a loop.

Syntax:
map(function, iterable)
- `function`: The function that is applied to each element in the iterable.
- `iterable`: The iterable (such as a list or tuple) whose elements are passed to the function.

Usage:
- It’s often used for transforming or modifying elements of a sequence in a clean and efficient way, especially when the transformation can be expressed as a single function.
# Key Points:
- Returns an iterator: The result is a map object, which needs to be converted to a list (or another iterable type) to view the results.
- Memory efficient: Since it doesn't generate a full list upfront, `map()` is memory-efficient for large data.


```python
#Example:
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))

```

```python
#Example with Multiple Iterables:

a = [1, 2, 3]
b = [4, 5, 6]
result = map(lambda x, y: x + y, a, b)
print(list(result))
```

Que10: What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 Ans10:

 In Python, `map()`, `reduce()`, and `filter()` are functional programming functions that operate on iterables.

1. `map(function, iterable)`
- Applies a given function to each element of an iterable and returns a new iterable (map object).

```python
#Example: Squaring each number in the list.

numbers = [47, 11, 42, 13]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [2209, 121, 1764, 169]
```

2. `filter(function, iterable)`
- it filters elements based on a given function that returns `True` or `False`.

```python
#Example: Filtering even numbers from the list.

numbers = [47, 11, 42, 13]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)
```

3. `reduce(function, iterable)`
- It reduces the iterable to a single accumulated value.
- It is available in the `functools` module.

```python
#Example: Finding the sum of all numbers in the list.

from functools import reduce
numbers = [47, 11, 42, 13]
sum_all = reduce(lambda x, y: x + y, numbers)
print(sum_all)
```

In Short what these functions basically do to a statement is:
'map()' - Applies a function to each element and returns a new iterable.
'filter()' - Selects elements based on a condition and returns a subset.
'reduce()' - Aggregates elements into a single value.

Que4: Write a Python function that checks if a given number is prime or not from 1 to 200.

Ans4: A prime number is a number that is divisible by itself and 1.

```python
#Ans4:
def is_prime(n):
    if n < 2:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to √n
        if n % i == 0:
            return False  # If divisible, not a prime
    return True  # Otherwise, it's prime

# Checking prime numbers from 1 to 200
primes = [num for num in range(1, 201) if is_prime(num)]
print("Prime numbers from 1 to 200:", primes)
```

Que5: Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
#Ans5:
def fibonacci(n):
    a, b = 0, 1  # First two Fibonacci numbers
    for i in range(n):
        yield a  # Yield the current number
        a, b = b, a + b  # Move to the next numbers

# Example usage
n = int(input("Enter the number of Fibonacci terms: "))
print("Fibonacci Sequence:", list(fibonacci(n)))  # Convert generator to list
```

Que6: Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
#Ans6:
def pow_gen(m):
    for exponent in range(m + 1):  # Loop from 0 to m
        yield 2 ** exponent  # Yield 2 raised to the power of exponent

max_exp = int(input("Enter the maximum exponent: "))
print("Powers of 2:", list(pow_gen(max_exp)))  # Convert generator to list
```
